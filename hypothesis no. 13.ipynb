### Plotting the data so we can visualize the structure of data

```python
# Plotting Default Loan Count over Years
plt.figure(figsize=(9,3))
sns.lineplot(data=df, x='Year', y='default loan count', marker='o', color='blue', label='Default Loan Count')
plt.title('Default Loan Count Over the Years')
plt.xlabel('Year')
plt.ylabel('Default Loan Count')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Default Loan Amount over Years
plt.figure(figsize=(9,3))
sns.lineplot(data=df, x='Year', y='default loan amount', marker='o', color='green', label='Default Loan Amount')
plt.title('Default Loan Amount Over the Years')
plt.xlabel('Year')
plt.ylabel('Default Loan Amount (in USD)')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Stock market index over Years
plt.figure(figsize=(9,3))
sns.lineplot(data=df, x='Year', y='Stock market index', marker='o', color='red', label='Stock market index')
plt.title('Stock market index Over the Years')
plt.xlabel('Year')
plt.ylabel('Stock market index')
plt.grid(True)
plt.legend()
plt.show()
```

### Visualizing the relationship of data to each other

```python
# Plotting Default Loan Count vs Stock market index
plt.figure(figsize=(9,3))
plt.plot(df['default loan count'], df['Stock market index'], marker='o', color='blue', linestyle='-')
plt.title('Default Loan Count vs Stock market index')
plt.xlabel('Default Loan Count')
plt.ylabel('Stock market index')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Default Loan Amount vs Stock market index
plt.figure(figsize=(9,3))
plt.plot(df['default loan amount'], df['Stock market index'], marker='o', color='green', linestyle='-')
plt.title('Default Loan Amount vs Stock market index')
plt.xlabel('Default Loan Amount')
plt.ylabel('Stock market index')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Filtering the DataFrame for the years 2020 to 2023
df_filtered = df[df['Year'].between(2020, 2023)]

# Standardize the data
default_loan_amount_standardized = (df_filtered['default loan amount'] - df_filtered['default loan amount'].mean()) / df_filtered['default loan amount'].std()
default_loan_count_standardized = (df_filtered['default loan count'] - df_filtered['default loan count'].mean()) / df_filtered['default loan count'].std()
Stock_market_index_standardized = (df_filtered['Stock market index'] - df_filtered['Stock market index'].mean()) / df_filtered['Stock market index'].std()

# Plotting Default Loan Amount and Stock market index
plt.figure(figsize=(9,3))
plt.plot(df_filtered['Year'], default_loan_amount_standardized, marker='o', color='green', linestyle='-', label='Default Loan Amount (standardized)')
plt.plot(df_filtered['Year'], Stock_market_index_standardized, marker='o', color='red', linestyle='-', label='Stock market index (standardized)')
plt.title('Default Loan Amount and Stock market index')
plt.xlabel('Year')
plt.ylabel('Standardized Value')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Default Loan Count and Stock market index
plt.figure(figsize=(9,3))
plt.plot(df_filtered['Year'], default_loan_count_standardized, marker='o', color='blue', linestyle='-', label='Default Loan Count (standardized)')
plt.plot(df_filtered['Year'], Stock_market_index_standardized, marker='o', color='red', linestyle='-', label='Stock market index (standardized)')
plt.title('Default Loan Count and Stock market index')
plt.xlabel('Year')
plt.ylabel('Standardized Value')
plt.grid(True)
plt.legend()
plt.show()
```

### Calculating Pearson's Correlation coefficient to see the strength of relationship

```python
df[['Stock market index','default loan amount','default loan count']].corr()
```

- Correlation '-0.230589' between 'Stock market index' and 'default loan amount' and Correlation '-0.236112' between 'Stock market index' and 'default loan count' suggest that the relationships between them are negative but not significantly strong.

# hypothesis testing: p-test

Step1: Define null and alternative hypothesis

- Null Hypothesis(H0): Stock market index does not have a positive relationship with non-performing loans or default                                   loans in Nepal
- Alternative Hypothesis(H1): Stock market index has a positive relationship with non-performing loans or default loans in Nepal

Step 2: Decide the significance level

α = 0.05

Step 3: Calculation of p-value

```python
from scipy.stats import spearmanr

# Dropping rows with NaN values for correlation analysis
df_cleaned_amount = df.dropna(subset=['Stock market index', 'default loan amount'])
df_cleaned_count = df.dropna(subset=['Stock market index', 'default loan count'])

# Calculating Spearmanr correlation coefficient and p-value for default loan amount
correlation_amount, p_value_amount = spearmanr(df_cleaned_amount['Stock market index'], df_cleaned_amount['default loan amount'])

# Calculating Spearmanr correlation coefficient and p-value for default loan count
correlation_count, p_value_count = spearmanr(df_cleaned_count['Stock market index'], df_cleaned_count['default loan count'])

# Printing the results
print("Correlation between Stock market index and Default Loan Amount:", correlation_amount)
print("P-value for Default Loan Amount:", p_value_amount)

print("Correlation between Stock market index and Default Loan Count:", correlation_count)
print("P-value for Default Loan Count:", p_value_count)
```

Results Interpretation:
- Correlation Coefficient: The correlation between the Stock Market Index and both Default Loan Amount and Default Loan Count is approximately 0.20, indicating a weak positive relationship.

- P-Value: Both p-values are 0.8, which exceeds the common significance level of 0.05.

Conclusion:
- The data shows a weak positive relationship between the Stock Market Index and both Default Loan Amount and Default Loan Count. However, with p-values of 0.8, we fail to reject the null hypothesis, meaning there isn't enough evidence to say this relationship is statistically significant. In simple terms, while there is some connection, it isn’t strong enough to confidently state that stock market index levels affect default loans.
