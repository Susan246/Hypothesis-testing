### Visualizing the relationship of data to each other

```python
# Plotting Default Loan Count vs Per capita GDP Growth
plt.figure(figsize=(9,3))
plt.plot(df['default loan count'], df['Per capita GDP Growth'], marker='o', color='blue', linestyle='-')
plt.title('Default Loan Count vs Per capita GDP Growth')
plt.xlabel('Default Loan Count')
plt.ylabel('Per capita GDP Growth')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Default Loan Amount vs Per capita GDP Growth
plt.figure(figsize=(9,3))
plt.plot(df['default loan amount'], df['Per capita GDP Growth'], marker='o', color='green', linestyle='-')
plt.title('Default Loan Amount vs Per capita GDP Growth')
plt.xlabel('Default Loan Amount')
plt.ylabel('Per capita GDP Growth')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Filtering the DataFrame for the years 2020 to 2023
df_filtered = df[df['Year'].between(2020, 2023)]

# Standardize the data
default_loan_amount_standardized = (df_filtered['default loan amount'] - df_filtered['default loan amount'].mean()) / df_filtered['default loan amount'].std()
default_loan_count_standardized = (df_filtered['default loan count'] - df_filtered['default loan count'].mean()) / df_filtered['default loan count'].std()
Per_capita_GDP_Growth_standardized = (df_filtered['Per capita GDP Growth'] - df_filtered['Per capita GDP Growth'].mean()) / df_filtered['Per capita GDP Growth'].std()

# Plotting Default Loan Amount and Per capita GDP Growth
plt.figure(figsize=(9,3))
plt.plot(df_filtered['Year'], default_loan_amount_standardized, marker='o', color='green', linestyle='-', label='Default Loan Amount (standardized)')
plt.plot(df_filtered['Year'], Per_capita_GDP_Growth_standardized, marker='o', color='red', linestyle='-', label='Per capita GDP Growth (standardized)')
plt.title('Default Loan Amount and Per capita GDP Growth')
plt.xlabel('Year')
plt.ylabel('Standardized Value')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Default Loan Count and Per capita GDP Growth
plt.figure(figsize=(9,3))
plt.plot(df_filtered['Year'], default_loan_count_standardized, marker='o', color='blue', linestyle='-', label='Default Loan Count (standardized)')
plt.plot(df_filtered['Year'], Per_capita_GDP_Growth_standardized, marker='o', color='red', linestyle='-', label='Per capita GDP Growth (standardized)')
plt.title('Default Loan Count and Per capita GDP Growth')
plt.xlabel('Year')
plt.ylabel('Standardized Value')
plt.grid(True)
plt.legend()
plt.show()
```

# hypothesis testing: p-test

Step1: Define null and alternative hypothesis

- Null Hypothesis(H0): GDP per Capita Growth does not have a positive relationship with non-performing loans or default                                   loans in Nepal
- Alternative Hypothesis(H1) : GDP per Capita Growth has a positive relationship with non-performing loans or default loans in Nepal

Step 2: Decide the significance level
    
Î± = 0.05

Step 3: Calculation of p-value

```python
from scipy.stats import spearmanr

# Dropping rows with NaN values for correlation analysis
df_cleaned_amount = df.dropna(subset=['Per capita GDP Growth', 'default loan amount'])
df_cleaned_count = df.dropna(subset=['Per capita GDP Growth', 'default loan count'])

# Calculating Spearmanr correlation coefficient and p-value for default loan amount
correlation_amount, p_value_amount = spearmanr(df_cleaned_amount['Per capita GDP Growth'], df_cleaned_amount['default loan amount'])

# Calculating Spearmanr correlation coefficient and p-value for default loan count
correlation_count, p_value_count = spearmanr(df_cleaned_count['Per capita GDP Growth'], df_cleaned_count['default loan count'])

# Printing the results
print("Correlation between Per capita GDP Growth and Default Loan Amount:", correlation_amount)
print("P-value for Default Loan Amount:", p_value_amount)

print("Correlation between Per capita GDP Growth and Default Loan Count:", correlation_count)
print("P-value for Default Loan Count:", p_value_count)
```

Results Interpretation:

- Correlation Coefficient: The correlation between Per Capita GDP Growth and Default Loan Amount, as well as Default Loan Count, is approximately 0.40, indicating a moderate positive relationship.

- P-Value: Both p-values are 0.6, which exceeds the common significance level of 0.05.

Conclusion:

- The data shows a moderate positive relationship between Per Capita GDP Growth and both Default Loan Amount and Default Loan Count. However, with p-values of 0.6, we fail to reject the null hypothesis, meaning there isn't enough evidence to say this relationship is statistically significant. In simple terms, while there is some connection, it isnâ€™t strong enough to confidently state that GDP growth affects default loans.
