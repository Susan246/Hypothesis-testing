### Plotting the data so we can visualize the structure of data

```python
# Plotting Default Loan Count over Years
plt.figure(figsize=(9,3))
sns.lineplot(data=df, x='Year', y='default loan count', marker='o', color='blue', label='Default Loan Count')
plt.title('Default Loan Count Over the Years')
plt.xlabel('Year')
plt.ylabel('Default Loan Count')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Default Loan Amount over Years
plt.figure(figsize=(9,3))
sns.lineplot(data=df, x='Year', y='default loan amount', marker='o', color='green', label='Default Loan Amount')
plt.title('Default Loan Amount Over the Years')
plt.xlabel('Year')
plt.ylabel('Default Loan Amount (in USD)')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Per capita GDP Growth over Years
plt.figure(figsize=(9,3))
sns.lineplot(data=df, x='Year', y='GDP By industries', marker='o', color='red', label='GDP By industries')
plt.title('GDP By industries Over the Years')
plt.xlabel('Year')
plt.ylabel('GDP By industries (%)')
plt.grid(True)
plt.legend()
plt.show()
```

### Visualizing the relationship of data to each other

```python
# Plotting Default Loan Count vs GDP By industries
plt.figure(figsize=(9,3))
plt.plot(df['default loan count'], df['GDP By industries'], marker='o', color='blue', linestyle='-')
plt.title('Default Loan Count vs GDP By industries')
plt.xlabel('Default Loan Count')
plt.ylabel('GDP By industries')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Default Loan Amount vs GDP By industries
plt.figure(figsize=(9,3))
plt.plot(df['default loan amount'], df['GDP By industries'], marker='o', color='green', linestyle='-')
plt.title('Default Loan Amount vs GDP By industries')
plt.xlabel('Default Loan Amount')
plt.ylabel('Per capita GDP Growth')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Filtering the DataFrame for the years 2020 to 2023
df_filtered = df[df['Year'].between(2020, 2023)]

# Standardize the data
default_loan_amount_standardized = (df_filtered['default loan amount'] - df_filtered['default loan amount'].mean()) / df_filtered['default loan amount'].std()
default_loan_count_standardized = (df_filtered['default loan count'] - df_filtered['default loan count'].mean()) / df_filtered['default loan count'].std()
GDP_By_industries_standardized = (df_filtered['GDP By industries'] - df_filtered['GDP By industries'].mean()) / df_filtered['GDP By industries'].std()

# Plotting Default Loan Amount and GDP By industries
plt.figure(figsize=(9,3))
plt.plot(df_filtered['Year'], default_loan_amount_standardized, marker='o', color='green', linestyle='-', label='Default Loan Amount (standardized)')
plt.plot(df_filtered['Year'], GDP_By_industries_standardized, marker='o', color='red', linestyle='-', label='GDP By industries (standardized)')
plt.title('Default Loan Amount and GDP By industries')
plt.xlabel('Year')
plt.ylabel('Standardized Value')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Default Loan Count and GDP By industries
plt.figure(figsize=(9,3))
plt.plot(df_filtered['Year'], default_loan_count_standardized, marker='o', color='blue', linestyle='-', label='Default Loan Count (standardized)')
plt.plot(df_filtered['Year'], GDP_By_industries_standardized, marker='o', color='red', linestyle='-', label='GDP By industries (standardized)')
plt.title('Default Loan Count and GDP By industries')
plt.xlabel('Year')
plt.ylabel('Standardized Value')
plt.grid(True)
plt.legend()
plt.show()
```

### Calculating Pearson's Correlation coefficient to see the strength of relationship

```python
df[['GDP By industries','default loan amount','default loan count']].corr()
```

- The correlation between GDP by industries and default loan amount is 0.547, suggesting a moderate positive relationship.
- The correlation between GDP by industries and default loan count is 0.548, also indicating a moderate positive relationship.

These correlations suggest a positive relationship between GDP by industries and both default loan amount and default loan count, with the relationships being moderately strong.

# hypothesis testing: p-test

Step1: Define null and alternative hypothesis

- Null Hypothesis(H0): GDP by industries  does not have a positive relationship with non-performing loans or default                                   loans in Nepal
- Alternative Hypothesis(H1) : GDP by industries  has a positive relationship with non-performing loans or default loans in Nepal

Step 2: Decide the significance level
    
Î± = 0.05

Step 3: Calculation of p-value

```python
from scipy.stats import spearmanr

# Dropping rows with NaN values for correlation analysis
df_cleaned_amount = df.dropna(subset=['GDP By industries', 'default loan amount'])
df_cleaned_count = df.dropna(subset=['GDP By industries', 'default loan count'])

# Calculating Spearmanr correlation coefficient and p-value for default loan amount
correlation_amount, p_value_amount = spearmanr(df_cleaned_amount['GDP By industries'], df_cleaned_amount['default loan amount'])

# Calculating Spearmanr correlation coefficient and p-value for default loan count
correlation_count, p_value_count = spearmanr(df_cleaned_count['GDP By industries'], df_cleaned_count['default loan count'])

# Printing the results
print("Correlation between GDP By industries and Default Loan Amount:", correlation_amount)
print("P-value for Default Loan Amount:", p_value_amount)

print("Correlation between GDP By industries and Default Loan Count:", correlation_count)
print("P-value for Default Loan Count:", p_value_count)
```

Results Interpretation:

- Correlation Coefficient: The correlation between GDP by industries and Default Loan Amount, as well as Default Loan Count, is approximately 0.60, indicating a moderately strong positive relationship.

- P-Value: Both p-values are around 0.40, which exceeds the common significance level of 0.05.

Conclusion:

- The data shows a moderately strong positive relationship between GDP by industries and both Default Loan Amount and Default Loan Count. However, with p-values of approximately 0.40, we fail to reject the null hypothesis. This means there isn't enough evidence to claim that these relationships are statistically significant. In simple terms, while there is a noticeable connection, it isn't strong enough to confidently state that GDP by industries has a significant impact on default loans.
