```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Plotting the data so we can visualize the structure of data

```python
# Plotting Default Loan Count over Years
plt.figure(figsize=(9,3))
sns.lineplot(data=df, x='Year', y='default loan count', marker='o', color='blue', label='Default Loan Count')
plt.title('Default Loan Count Over the Years')
plt.xlabel('Year')
plt.ylabel('Default Loan Count')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Default Loan Amount over Years
plt.figure(figsize=(9,3))
sns.lineplot(data=df, x='Year', y='default loan amount', marker='o', color='green', label='Default Loan Amount')
plt.title('Default Loan Amount Over the Years')
plt.xlabel('Year')
plt.ylabel('Default Loan Amount (in USD)')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Average consumer inflation over Years
plt.figure(figsize=(9,3))
sns.lineplot(data=df, x='Year', y='Inflation', marker='o', color='red', label='Average consumer inflation')
plt.title('Average consumer inflation Over the Years')
plt.xlabel('Year')
plt.ylabel('Average consumer inflation (%)')
plt.grid(True)
plt.legend()
plt.show()
```

### Visualizing the relationship of data to each other

```python
# Plotting Default Loan Count vs Average consumer inflation
plt.figure(figsize=(9,3))
plt.plot(df['default loan count'], df['Inflation'], marker='o', color='blue', linestyle='-')
plt.title('Default Loan Count vs Average consumer inflation')
plt.xlabel('Default Loan Count')
plt.ylabel('Average consumer inflation')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Default Loan Amount vs Average consumer inflation
plt.figure(figsize=(9,3))
plt.plot(df['default loan amount'], df['Inflation'], marker='o', color='green', linestyle='-')
plt.title('Default Loan Amount vs Average consumer inflation')
plt.xlabel('Default Loan Amount')
plt.ylabel('Average consumer inflation')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Filtering the DataFrame for the years 2020 to 2023
df_filtered = df[df['Year'].between(2020, 2023)]

# Standardize the data
default_loan_amount_standardized = (df_filtered['default loan amount'] - df_filtered['default loan amount'].mean()) / df_filtered['default loan amount'].std()
default_loan_count_standardized = (df_filtered['default loan count'] - df_filtered['default loan count'].mean()) / df_filtered['default loan count'].std()
Inflation_standardized = (df_filtered['Inflation'] - df_filtered['Inflation'].mean()) / df_filtered['Inflation'].std()

# Plotting Default Loan Amount and Average consumer inflation
plt.figure(figsize=(9,3))
plt.plot(df_filtered['Year'], default_loan_amount_standardized, marker='o', color='green', linestyle='-', label='Default Loan Amount (standardized)')
plt.plot(df_filtered['Year'], Inflation_standardized, marker='o', color='red', linestyle='-', label='Average consumer inflation (standardized)')
plt.title('Default Loan Amount and Average consumer inflation')
plt.xlabel('Year')
plt.ylabel('Standardized Value')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Default Loan Count and Average consumer inflation
plt.figure(figsize=(9,3))
plt.plot(df_filtered['Year'], default_loan_count_standardized, marker='o', color='blue', linestyle='-', label='Default Loan Count (standardized)')
plt.plot(df_filtered['Year'], Inflation_standardized, marker='o', color='red', linestyle='-', label='Average consumer inflation (standardized)')
plt.title('Default Loan Count and Average consumer inflation')
plt.xlabel('Year')
plt.ylabel('Standardized Value')
plt.grid(True)
plt.legend()
plt.show()
```

### Calculating Pearson's Correlation coefficient to see the strength of relationship

```python
df[['Inflation','default loan amount','default loan count']].corr()
```

- The correlation between Inflation and Default Loan Amount is approximately 0.87, and the correlation between Inflation and Default Loan Count is also around 0.87. These values suggest a strong positive relationship between inflation and both default loan amount and default loan count. This indicates that as inflation increases, default loans tend to increase as well.

# hypothesis testing: p-test

Step1: Define null and alternative hypothesis

- Null Hypothesis(H0):  Average consumer inflation does not have a positive relationship with non-performing loans or default                                   loans in Nepal
- Alternative Hypothesis(H1) :  Average consumer inflation has a positive relationship with non-performing loans or default loans in Nepal

Step 2: Decide the significance level
    
Î± = 0.05

Step 3: Calculation of p-value

```python
from scipy.stats import spearmanr

# Dropping rows with NaN values for correlation analysis
df_cleaned_amount = df.dropna(subset=['Inflation', 'default loan amount'])
df_cleaned_count = df.dropna(subset=['Inflation', 'default loan count'])

# Calculating Spearmanr correlation coefficient and p-value for default loan amount
correlation_amount, p_value_amount = spearmanr(df_cleaned_amount['Inflation'], df_cleaned_amount['default loan amount'])

# Calculating Spearmanr correlation coefficient and p-value for default loan count
correlation_count, p_value_count = spearmanr(df_cleaned_count['Inflation'], df_cleaned_count['default loan count'])

# Printing the results
print("Correlation between  Average consumer inflation and Default Loan Amount:", correlation_amount)
print("P-value for Default Loan Amount:", p_value_amount)

print("Correlation between  Average consumer inflation and Default Loan Count:", correlation_count)
print("P-value for Default Loan Count:", p_value_count)
```

Results Interpretation:

- Correlation Coefficient: The correlation between Average Consumer Inflation and both Default Loan Amount and Default Loan Count is approximately 0.60. This indicates a moderately strong positive relationship, suggesting that higher inflation is associated with higher default loan amounts and counts.

- P-Value: Both p-values are approximately 0.40, which is greater than the common significance level of 0.05.

Conclusion:

- Although the correlations are moderately strong, the p-values are greater than 0.05, meaning we fail to reject the null hypothesis. This indicates that there is not enough statistical evidence to conclude that the relationship between Average Consumer Inflation and default loans is statistically significant. In simple terms, while there is a noticeable positive connection, it is not significant enough to confidently assert a meaningful impact of inflation on default loans.
