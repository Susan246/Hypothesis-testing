```python
# Filtering the DataFrame for the years 2020 to 2023
df_filtered = df[df['Year'].between(2020, 2023)]

# Standardize the data
default_loan_amount_standardized = (df_filtered['default loan amount'] - df_filtered['default loan amount'].mean()) / df_filtered['default loan amount'].std()
default_loan_count_standardized = (df_filtered['default loan count'] - df_filtered['default loan count'].mean()) / df_filtered['default loan count'].std()
deposite_standardized = (df_filtered['Deposite'] - df_filtered['Deposite'].mean()) / df_filtered['Deposite'].std()
credit_standardized = (df_filtered['Credit'] - df_filtered['Credit'].mean()) / df_filtered['Credit'].std()

# Plotting Default Loan Amount and Deposite
plt.figure(figsize=(9,3))
plt.plot(df_filtered['Year'], default_loan_amount_standardized, marker='o', color='green', linestyle='-', label='Default Loan Amount (standardized)')
plt.plot(df_filtered['Year'], deposite_standardized, marker='o', color='red', linestyle='-', label='Deposite (standardized)')
plt.title('Default Loan Amount and Deposite')
plt.xlabel('Year')
plt.ylabel('Standardized Value')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Default Loan Count and Deposite
plt.figure(figsize=(9,3))
plt.plot(df_filtered['Year'], default_loan_count_standardized, marker='o', color='blue', linestyle='-', label='Default Loan Count (standardized)')
plt.plot(df_filtered['Year'], deposite_standardized, marker='o', color='red', linestyle='-', label='Deposite (standardized)')
plt.title('Default Loan Count and Deposite')
plt.xlabel('Year')
plt.ylabel('Standardized Value')
plt.grid(True)
plt.legend()
plt.show()


# Plotting Default Loan Amount and Credit
plt.figure(figsize=(9,3))
plt.plot(df_filtered['Year'], default_loan_amount_standardized, marker='o', color='green', linestyle='-', label='Default Loan Amount (standardized)')
plt.plot(df_filtered['Year'], credit_standardized, marker='o', color='red', linestyle='-', label='Credit (standardized)')
plt.title('Default Loan Amount and Credit')
plt.xlabel('Year')
plt.ylabel('Standardized Value')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Default Loan Count and Credit
plt.figure(figsize=(9,3))
plt.plot(df_filtered['Year'], default_loan_count_standardized, marker='o', color='blue', linestyle='-', label='Default Loan Count (standardized)')
plt.plot(df_filtered['Year'], credit_standardized, marker='o', color='red', linestyle='-', label='Credit (standardized)')
plt.title('Default Loan Count and Credit')
plt.xlabel('Year')
plt.ylabel('Standardized Value')
plt.grid(True)
plt.legend()
plt.show()
```

# hypothesis testing:
## p-test of Deposite and default loan

Step1: Define null and alternative hypothesis

- Null Hypothesis(H0): Deposite does not have a positive relationship with non-performing loans or default                                   loans in Nepal
- Alternative Hypothesis(H1): Deposite has a positive relationship with non-performing loans or default loans in Nepal

Step 2: Decide the significance level

α = 0.05

Step 3: Calculation of p-value

```python
from scipy.stats import spearmanr

# Dropping rows with NaN values for correlation analysis
df_cleaned_amount = df.dropna(subset=['Deposite', 'default loan amount'])
df_cleaned_count = df.dropna(subset=['Deposite', 'default loan count'])

# Calculating Spearmanr correlation coefficient and p-value for default loan amount
correlation_amount, p_value_amount = spearmanr(df_cleaned_amount['Deposite'], df_cleaned_amount['default loan amount'])

# Calculating Spearmanr correlation coefficient and p-value for default loan count
correlation_count, p_value_count = spearmanr(df_cleaned_count['Deposite'], df_cleaned_count['default loan count'])

# Printing the results
print("Correlation between Deposite and Default Loan Amount:", correlation_amount)
print("P-value for Default Loan Amount:", p_value_amount)

print("Correlation between Deposite and Default Loan Count:", correlation_count)
print("P-value for Default Loan Count:", p_value_count)
```

Interpretation of Results:

- P-Value: Both p-values (0.0) are less than the common significance level of 0.05. Therefore, we reject the null hypothesis.

- Correlation Coefficient: The correlation coefficient of 1.0 indicates a perfect positive correlation between Deposits and both Default Loan Amount and Default Loan Count. Given the low p-values, this correlation is statistically significant.

Conclusion Based on Results:

- Reject Null Hypothesis: Since the p-values are less than 0.05, we have enough statistical evidence to reject the null hypothesis.

- Statistically Significant Relationship: The perfect positive correlation observed is statistically significant, as indicated by the low p-values.

Final Conclusion:
- Based on the p-values, we reject the null hypothesis that the Deposits do not have a positive relationship with non-performing loans or default loans in Nepal. This means that, given the current data, we have sufficient evidence to conclude that there is a statistically significant positive relationship between Deposits and non-performing or default loans. It is important to note that rejecting the null hypothesis indicates strong evidence for the relationship.
