### Visualizing the graph of Inflation vs Growth in Salary and Wage index

```python
# Plotting Inflation vs Growth in Salary and Wage index
plt.figure(figsize=(9,3))
plt.plot(df['Inflation'], df['Growth in salary and wage index'], marker='o', color='blue', linestyle='-')
plt.title('Inflation vs Growth in Salary and Wage index')
plt.xlabel('Inflation')
plt.ylabel('Growth in salary and wage index')
plt.grid(True)
plt.legend()
plt.show()
```

- we are going to standardize the data to ensure that variables Inflation and Growth in Salary and Wage Index are on the same scale, making it easier to accurately assess their relationship with default loans.

```python
from sklearn.preprocessing import StandardScaler

# Standardizie the data
scaler = StandardScaler()
df[['Inflation', 'Growth in salary and wage index']] = scaler.fit_transform(df[['Inflation', 'Growth in salary and wage index']])

# Plotting the standardized data
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Inflation'], label='Inflation (Standardized)')
plt.plot(df['Year'], df['Growth in salary and wage index'], label='Growth in Salary and Wage Index (Standardized)')
plt.xlabel('Year')
plt.ylabel('Standardized Value')
plt.title('Standardized Inflation vs Growth in Salary and Wage Index')
plt.legend()
plt.grid(True)
plt.show()
```

### We are going to make 1 column from 'Inflation' and 'Growth in Salary and Wage index' using PCA so we can test hypothesis

- Standarizing the data make PCA perform well and we have already done that

```python
from sklearn.decomposition import PCA

# Drop rows with NaN values
df= df.dropna(subset=['Inflation', 'Growth in salary and wage index'])

# Applying PCA
pca=PCA()
X_pca=pca.fit_transform(df[['Inflation','Growth in salary and wage index']])

# Eigenvalues 
eigenvalues = pca.explained_variance_


# Printing eigenvalues
print("Eigenvalues:", eigenvalues)

# Finding the index of the highest eigenvalue
best_component_index = np.argmax(eigenvalues)

print(f"The best principal component is Component {best_component_index + 1} with eigenvalue {eigenvalues[best_component_index]}")

# Selecting the best Principle Component
best_component = X_pca[:, best_component_index]

# Adding the best principal component to the DataFrame
df['pca'] = best_component

print(df)
```

- We have successfully transform 2 columns; 'Inflation' and 'Growth in salary and wage index' into 1 with the help of PCA

# hypothesis testing: p-test

Step1: Define null and alternative hypothesis

- Null Hypothesis(H0): 'Inflation vs Growth in Salary and Wage index' does not have a positive relationship with non-performing loans or default loans in Nepal
- Alternative Hypothesis(H1): 'Inflation vs Growth in Salary and Wage index' has a positive relationship with non-performing loans or default loans in Nepal

Step 2: Decide the significance level

 = 0.05

Step 3: Calculation of p-value

```python
from scipy.stats import spearmanr

# Dropping rows with NaN values for correlation analysis
df_cleaned_amount = df.dropna(subset=['pca', 'default loan amount'])
df_cleaned_count = df.dropna(subset=['pca', 'default loan count'])

# Calculating Spearmanr correlation coefficient and p-value for default loan amount
correlation_amount, p_value_amount = spearmanr(df_cleaned_amount['pca'], df_cleaned_amount['default loan amount'])

# Calculating Spearmanr correlation coefficient and p-value for default loan count
correlation_count, p_value_count = spearmanr(df_cleaned_count['pca'], df_cleaned_count['default loan count'])

# Printing the results
print("Correlation between 'Inflation vs Growth in Salary and Wage index'  and Default Loan Amount:", correlation_amount)
print("P-value for Default Loan Amount:", p_value_amount)

print("Correlation between 'Inflation vs Growth in Salary and Wage index'  and Default Loan Count:", correlation_count)
print("P-value for Default Loan Count:", p_value_count)
```

Results Interpretation:

- Correlation Coefficient: The correlation between 'Inflation vs Growth in Salary and Wage Index' and both Default Loan Amount and Default Loan Count is approximately 0.60, indicating a moderate positive relationship.

- P-Value: Both p-values are 0.40, which exceeds the common significance level of 0.05.

Conclusion:

- The data shows a moderate positive relationship between 'Inflation vs Growth in Salary and Wage Index' and both Default Loan Amount and Default Loan Count. However, with p-values of 0.40, we fail to reject the null hypothesis, meaning there isn't enough evidence to say this relationship is statistically significant. In simple terms, while there is some connection, it isnt strong enough to confidently state that 'Inflation vs Growth in Salary and Wage Index' affects default loans.
