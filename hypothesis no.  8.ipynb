### Plotting the graph so we can visualize the structure of data

```python
# Plotting Default Loan Count over Years
plt.figure(figsize=(9,3))
sns.lineplot(data=df, x='Year', y='default loan count', marker='o', color='blue', label='Default Loan Count')
plt.title('Default Loan Count Over the Years')
plt.xlabel('Year')
plt.ylabel('Default Loan Count')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Default Loan Amount over Years
plt.figure(figsize=(9,3))
sns.lineplot(data=df, x='Year', y='default loan amount', marker='o', color='green', label='Default Loan Amount')
plt.title('Default Loan Amount Over the Years')
plt.xlabel('Year')
plt.ylabel('Default Loan Amount (in USD)')
plt.grid(True)
plt.legend()
plt.show()

# Plotting  Poverty rate
plt.figure(figsize=(9,3))
sns.lineplot(data=df, x='Year', y='Poverty rate', marker='o', color='red', label='Poverty rate')
plt.title('Poverty rate Over the Years')
plt.xlabel('Year')
plt.ylabel('Poverty rate (%)')
plt.grid(True)
plt.legend()
plt.show()
```

### Visualizing the relationship of data to each other

```python
# Plotting Default Loan Count vs Poverty rate
plt.figure(figsize=(9,3))
plt.plot(df['default loan count'], df['Poverty rate'], marker='o', color='blue', linestyle='-')
plt.title('Default Loan Count vs Poverty rate')
plt.xlabel('Default Loan Count')
plt.ylabel('Poverty rate')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Default Loan Amount vs Poverty rate
plt.figure(figsize=(9,3))
plt.plot(df['default loan amount'], df['Poverty rate'], marker='o', color='green', linestyle='-')
plt.title('Default Loan Amount vs Poverty rate')
plt.xlabel('Default Loan Amount')
plt.ylabel('Poverty rate')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Filtering the DataFrame for the years 2020 to 2023
df_filtered = df[df['Year'].between(2020, 2023)]

# Standardize the data
default_loan_amount_standardized = (df_filtered['default loan amount'] - df_filtered['default loan amount'].mean()) / df_filtered['default loan amount'].std()
default_loan_count_standardized = (df_filtered['default loan count'] - df_filtered['default loan count'].mean()) / df_filtered['default loan count'].std()
actual_growth_rate_standardized = (df_filtered['Poverty rate'] - df_filtered['Poverty rate'].mean()) / df_filtered['Poverty rate'].std()

# Plotting Default Loan Amount and Poverty rate
plt.figure(figsize=(9,3))
plt.plot(df_filtered['Year'], default_loan_amount_standardized, marker='o', color='green', linestyle='-', label='Default Loan Amount (standardized)')
plt.plot(df_filtered['Year'], actual_growth_rate_standardized, marker='o', color='red', linestyle='-', label='Poverty rate (standardized)')
plt.title('Default Loan Amount and Poverty rate')
plt.xlabel('Year')
plt.ylabel('Standardized Value')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Default Loan Count and Poverty rate
plt.figure(figsize=(9,3))
plt.plot(df_filtered['Year'], default_loan_count_standardized, marker='o', color='blue', linestyle='-', label='Default Loan Count (standardized)')
plt.plot(df_filtered['Year'], actual_growth_rate_standardized, marker='o', color='red', linestyle='-', label='Poverty rate (standardized)')
plt.title('Default Loan Count and Poverty rate')
plt.xlabel('Year')
plt.ylabel('Standardized Value')
plt.grid(True)
plt.legend()
plt.show()
```

### Calculating Pearson's Correlation coefficient to see the strength of relationship

```python
df[['Poverty rate','default loan amount','default loan count']].corr()
```

- The correlation between the Poverty Rate and Default Loan Amount is approximately 0.95, and the correlation between the Poverty Rate and Default Loan Count is also around 0.95. These values suggest a very strong positive relationship between the poverty rate and both default loan amount and default loan count. This indicates that as the poverty rate increases, default loans tend to increase significantly as well.

# hypothesis testing: p-test

Step1: Define null and alternative hypothesis

- Null Hypothesis(H0): Poverty rate does not have a positive relationship with non-performing loans or default loans in Nepal
- Alternative Hypothesis(H1) : Poverty rate inflation has a positive relationship with non-performing loans or default loans in Nepal

Step 2: Decide the significance level

Î± = 0.05

Step 3: Calculation of p-value

```python
from scipy.stats import spearmanr

# Dropping rows with NaN values for correlation analysis
df_cleaned_amount = df.dropna(subset=['Poverty rate', 'default loan amount'])
df_cleaned_count = df.dropna(subset=['Poverty rate', 'default loan count'])

# Calculating Spearmanr correlation coefficient and p-value for default loan amount
correlation_amount, p_value_amount = spearmanr(df_cleaned_amount['Poverty rate'], df_cleaned_amount['default loan amount'])

# Calculating Spearmanr correlation coefficient and p-value for default loan count
correlation_count, p_value_count = spearmanr(df_cleaned_count['Poverty rate'], df_cleaned_count['default loan count'])

# Printing the results
print("Correlation between  Poverty rate and Default Loan Amount:", correlation_amount)
print("P-value for Default Loan Amount:", p_value_amount)

print("Correlation between  Poverty rate and Default Loan Count:", correlation_count)
print("P-value for Default Loan Count:", p_value_count)
```

Results Interpretation:

- Correlation Coefficient: The correlation between Poverty Rate and both Default Loan Amount and Default Loan Count is approximately 0.74, indicating a strong positive relationship.

- P-Value: Both p-values are approximately 0.26, which is greater than the common significance level of 0.05.

Conclusion:
- Although the correlation is strong, the p-values are greater than 0.05, so we fail to reject the null hypothesis. This means there is not enough statistical evidence to conclude that the relationship between Poverty Rate and default loans is statistically significant. Thus, while there is a strong positive connection, it is not significant enough to confidently assert a meaningful impact of poverty on default loans.
