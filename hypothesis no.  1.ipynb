# hypothesis testing: p-test

Step1: Define null and alternative hypothesis

- Null Hypothesis(H0): Actual Growth Rate does not have a positive relationship with non-performing loans or default                                   loans in Nepal
- Alternative Hypothesis(H1): Actual Growth Rate has a positive relationship with non-performing loans or default loans in Nepal

Step 2: Decide the significance level
    
α = 0.05

Step 3: Calculation of p-value

```python
from scipy.stats import spearmanr

# Dropping rows with NaN values for correlation analysis
df_cleaned_amount = df.dropna(subset=['Actual growth rate', 'default loan amount'])
df_cleaned_count = df.dropna(subset=['Actual growth rate', 'default loan count'])

# Calculating Spearmanr correlation coefficient and p-value for default loan amount
correlation_amount, p_value_amount = spearmanr(df_cleaned_amount['Actual growth rate'], df_cleaned_amount['default loan amount'])

# Calculating Spearmanr correlation coefficient and p-value for default loan count
correlation_count, p_value_count = spearmanr(df_cleaned_count['Actual growth rate'], df_cleaned_count['default loan count'])

# Printing the results
print("Correlation between Actual Growth Rate and Default Loan Amount:", correlation_amount)
print("P-value for Default Loan Amount:", p_value_amount)

print("Correlation between Actual Growth Rate and Default Loan Count:", correlation_count)
print("P-value for Default Loan Count:", p_value_count)
```

Interpretation of Results:

- P-Value: Both p-values (0.6) are greater than the common significance level of 0.05. Therefore, we fail to reject the null hypothesis.

- Correlation Coefficient: The correlation coefficient of approximately 0.4 indicates a moderate positive correlation between the Actual Growth Rate and both Default Loan Amount and Default Loan Count. However, given the high p-values, this correlation is not statistically significant.

Conclusion Based on Results:
- Fail to Reject Null Hypothesis: Since the p-values are greater than 0.05, we do not have enough statistical evidence to reject the null hypothesis.

- No Statistically Significant Relationship: The moderate positive correlation observed is not statistically significant, as indicated by the high p-values.

Final Conclusion:
- Based on the p-values, we fail to reject the null hypothesis that the Actual Growth Rate does not have a positive relationship with non-performing loans or default loans in Nepal. This means that, given the current data, we do not have sufficient evidence to conclude that there is a statistically significant positive relationship between Actual Growth Rate and non-performing or default loans. It is important to note that failing to reject the null hypothesis does not confirm it; it simply indicates that the data does not provide strong enough evidence against it.
