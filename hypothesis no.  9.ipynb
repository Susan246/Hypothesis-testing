# hypothesis testing: p-test

Step1: Define null and alternative hypothesis

Null Hypothesis(H0): Inequality Rate does not have a positive relationship with non-performing loans or default loans in Nepal
Alternative Hypothesis(H1): Inequality Rate has a positive relationship with non-performing loans or default loans in Nepal

Step 2: Decide the significance level

α = 0.05

Step 3: Calculation of p-value

```python
from scipy.stats import spearmanr

# Dropping rows with NaN values for correlation analysis
df_cleaned_amount = df.dropna(subset=['Inequality rate', 'default loan amount'])
df_cleaned_count = df.dropna(subset=['Inequality rate', 'default loan count'])

# Calculating Spearmanr correlation coefficient and p-value for default loan amount
correlation_amount, p_value_amount = spearmanr(df_cleaned_amount['Inequality rate'], df_cleaned_amount['default loan amount'])

# Calculating Spearmanr correlation coefficient and p-value for default loan count
correlation_count, p_value_count = spearmanr(df_cleaned_count['Inequality rate'], df_cleaned_count['default loan count'])

# Printing the results
print("Correlation between Inequality rate and Default Loan Amount:", correlation_amount)
print("P-value for Default Loan Amount:", p_value_amount)

print("Correlation between Inequality rate and Default Loan Count:", correlation_count)
print("P-value for Default Loan Count:", p_value_count)
```

Interpretation of Results:

- P-Value: Both p-values (0.0) are less than the common significance level of 0.05. Therefore, we reject the null hypothesis.

- Correlation Coefficient: The correlation coefficient of 1.0 indicates a perfect positive correlation between the Inequality Rate and both Default Loan Amount and Default Loan Count. Given the low p-values, this correlation is statistically significant.

Conclusion:
- The data shows a perfect positive relationship between the Inequality Rate and both Default Loan Amount and Default Loan Count. With p-values of 0.0, we reject the null hypothesis, meaning there is enough evidence to say this relationship is statistically significant. In simple terms, there is a strong connection, and we can confidently state that inequality rate affects default loans.
