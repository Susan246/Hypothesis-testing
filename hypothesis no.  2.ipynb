### Plotting the data so we can visualize the structure of data


```python

# Plotting Default Loan Count over Years
plt.figure(figsize=(9,3))
sns.lineplot(data=df, x='Year', y='default loan count', marker='o', color='blue', label='Default Loan Count')
plt.title('Default Loan Count Over the Years')
plt.xlabel('Year')
plt.ylabel('Default Loan Count')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Default Loan Amount over Years
plt.figure(figsize=(9,3))
sns.lineplot(data=df, x='Year', y='default loan amount', marker='o', color='green', label='Default Loan Amount')
plt.title('Default Loan Amount Over the Years')
plt.xlabel('Year')
plt.ylabel('Default Loan Amount (in USD)')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Forecasted Growth rate over Years
plt.figure(figsize=(9,3))
sns.lineplot(data=df, x='Year', y='Forecasted Growth rate', marker='o', color='red', label='Forecasted Growth rate')
plt.title('Forecasted Growth rate Over the Years')
plt.xlabel('Year')
plt.ylabel('Forecasted Growth rate (%)')
plt.grid(True)
plt.legend()
plt.show()
```

### Visualizing the relationship of data to each other

```python
# Plotting Default Loan Count vs Forecasted Growth rate
plt.figure(figsize=(9,3))
plt.plot(df['default loan count'], df['Forecasted Growth rate'], marker='o', color='blue', linestyle='-')
plt.title('Default Loan Count vs Forecasted Growth rate')
plt.xlabel('Default Loan Count')
plt.ylabel('Forecasted Growth rate')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Default Loan Amount vs Forecasted Growth rate
plt.figure(figsize=(9,3))
plt.plot(df['default loan amount'], df['Forecasted Growth rate'], marker='o', color='green', linestyle='-')
plt.title('Default Loan Amount vs Forecasted Growth rate')
plt.xlabel('Default Loan Amount')
plt.ylabel('Forecasted Growth rate')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Filtering the DataFrame for the years 2020 to 2023
df_filtered = df[df['Year'].between(2020, 2024)]

# Standardize the data
default_loan_amount_standardized = (df_filtered['default loan amount'] - df_filtered['default loan amount'].mean()) / df_filtered['default loan amount'].std()
default_loan_count_standardized = (df_filtered['default loan count'] - df_filtered['default loan count'].mean()) / df_filtered['default loan count'].std()
Forecasted_Growth_rate_standardized = (df_filtered['Forecasted Growth rate'] - df_filtered['Forecasted Growth rate'].mean()) / df_filtered['Forecasted Growth rate'].std()

# Plotting Default Loan Amount and Forecasted Growth rate
plt.figure(figsize=(9,3))
plt.plot(df_filtered['Year'], default_loan_amount_standardized, marker='o', color='green', linestyle='-', label='Default Loan Amount (standardized)')
plt.plot(df_filtered['Year'], Forecasted_Growth_rate_standardized, marker='o', color='red', linestyle='-', label='Forecasted Growth rate (standardized)')
plt.title('Default Loan Amount and Forecasted Growth rate')
plt.xlabel('Year')
plt.ylabel('Standardized Value')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Default Loan Count and Forecasted Growth rate
plt.figure(figsize=(9,3))
plt.plot(df_filtered['Year'], default_loan_count_standardized, marker='o', color='blue', linestyle='-', label='Default Loan Count (standardized)')
plt.plot(df_filtered['Year'], Forecasted_Growth_rate_standardized, marker='o', color='red', linestyle='-', label='Forecasted Growth rate (standardized)')
plt.title('Default Loan Count and Forecasted Growth rate')
plt.xlabel('Year')
plt.ylabel('Standardized Value')
plt.grid(True)
plt.legend()
plt.show()
```

### Calculating Pearson's Correlation coefficient to see the strength of relationship

```python
df[['Forecasted Growth rate','default loan amount','default loan count']].corr()
```

- Forecasted Growth Rate has a very weak negative correlation of -0.002632 with Default Loan Amount and a very weak positive correlation of 0.005656 with Default Loan Count. This indicates that the relationship between the Forecasted Growth Rate and both Default Loan Amount and Default Loan Count is negligible.

# hypothesis testing: p-test

Step1: Define null and alternative hypothesis

- Null Hypothesis(H0): Forecasted Growth Rate does not have a positive relationship with non-performing loans or default loans in Nepal
- Alternative Hypothesis(H1): Forecasted Growth Rate has a positive relationship with non-performing loans or default loans in Nepal

Step 2: Decide the significance level

Î± = 0.05

Step 3: Calculation of p-value

```python
from scipy.stats import spearmanr

# Dropping rows with NaN values for correlation analysis
df_cleaned_amount = df.dropna(subset=['Forecasted Growth rate', 'default loan amount'])
df_cleaned_count = df.dropna(subset=['Forecasted Growth rate', 'default loan count'])

# Calculating Spearmanr correlation coefficient and p-value for default loan amount
correlation_amount, p_value_amount = spearmanr(df_cleaned_amount['Forecasted Growth rate'], df_cleaned_amount['default loan amount'])

# Calculating Spearmanr correlation coefficient and p-value for default loan count
correlation_count, p_value_count = spearmanr(df_cleaned_count['Forecasted Growth rate'], df_cleaned_count['default loan count'])

# Printing the results
print("Correlation between Forecasted Growth Rate and Default Loan Amount:", correlation_amount)
print("P-value for Default Loan Amount:", p_value_amount)

print("Correlation between Forecasted Growth Rate and Default Loan Count:", correlation_count)
print("P-value for Default Loan Count:", p_value_count)
```


Results Interpretation:
- Correlation Coefficient: The correlation between the Forecasted Growth Rate and both Default Loan Amount and Default Loan Count is approximately -0.6, indicating a moderate negative relationship.

- P-Value: Both p-values are 0.285, which exceeds the common significance level of 0.05.

Conclusion:
- The data shows a moderate negative relationship between the Forecasted Growth Rate and both Default Loan Amount and Default Loan Count. However, with p-values of 0.285, we fail to reject the null hypothesis, meaning there isn't enough evidence to say this relationship is statistically significant. In simple terms, while there is some connection, it isnâ€™t strong enough to confidently state that forecasted growth rates affect default loans.
