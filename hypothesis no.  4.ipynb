```python
# Filtering the DataFrame for the years 2020 to 2023
df_filtered = df[df['Year'].between(2020, 2023)]

# Standardize the data
default_loan_amount_standardized = (df_filtered['default loan amount'] - df_filtered['default loan amount'].mean()) / df_filtered['default loan amount'].std()
default_loan_count_standardized = (df_filtered['default loan count'] - df_filtered['default loan count'].mean()) / df_filtered['default loan count'].std()
Nominal_GDP_standardized = (df_filtered['Nominal GDP'] - df_filtered['Nominal GDP'].mean()) / df_filtered['Nominal GDP'].std()
Real_GDP_standardized = (df_filtered['Real GDP'] - df_filtered['Real GDP'].mean()) / df_filtered['Real GDP'].std()

# Plotting Default Loan Amount and Nominal GDP
plt.figure(figsize=(9,3))
plt.plot(df_filtered['Year'], default_loan_amount_standardized, marker='o', color='green', linestyle='-', label='Default Loan Amount (standardized)')
plt.plot(df_filtered['Year'], Nominal_GDP_standardized, marker='o', color='red', linestyle='-', label='Nominal GDP (standardized)')
plt.title('Default Loan Amount and Nominal GDP')
plt.xlabel('Year')
plt.ylabel('Standardized Value')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Default Loan Count and  Nominal GDP
plt.figure(figsize=(9,3))
plt.plot(df_filtered['Year'], default_loan_count_standardized, marker='o', color='blue', linestyle='-', label='Default Loan Count (standardized)')
plt.plot(df_filtered['Year'], Nominal_GDP_standardized, marker='o', color='red', linestyle='-', label='Nominal GDP (standardized)')
plt.title('Default Loan Count and  Nominal GDP')
plt.xlabel('Year')
plt.ylabel('Standardized Value')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Default Loan Amount and Real GDP
plt.figure(figsize=(9,3))
plt.plot(df_filtered['Year'], default_loan_amount_standardized, marker='o', color='green', linestyle='-', label='Default Loan Amount (standardized)')
plt.plot(df_filtered['Year'], Real_GDP_standardized, marker='o', color='red', linestyle='-', label='Real GDP (standardized)')
plt.title('Default Loan Amount and Real GDP')
plt.xlabel('Year')
plt.ylabel('Standardized Value')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Default Loan Count and  Real GDP
plt.figure(figsize=(9,3))
plt.plot(df_filtered['Year'], default_loan_count_standardized, marker='o', color='blue', linestyle='-', label='Default Loan Count (standardized)')
plt.plot(df_filtered['Year'], Real_GDP_standardized, marker='o', color='red', linestyle='-', label='Real GDP (standardized)')
plt.title('Default Loan Count and Real GDP')
plt.xlabel('Year')
plt.ylabel('Standardized Value')
plt.grid(True)
plt.legend()
plt.show()
```

### Calculating Pearson's Correlation coefficient to see the strength of relationship

- The correlations between GDP (both nominal and real) and default loan amounts or counts are very high (around 0.93 to 0.96). This means that as GDP increases, both the amount and count of default loans also tend to increase.

- The default loan amount and count are almost perfectly correlated (about 0.9999), indicating that they move together closely.

# hypothesis testing: p-test
- Relationship of Nominal GDP and Default Loan

Step1: Define null and alternative hypothesis

- Null Hypothesis(H0): Nominal GDP does not have a positive relationship with non-performing loans or default loans in Nepal
- Alternative Hypothesis(H1) : Nominal GDP has a positive relationship with non-performing loans or default loans in Nepal

Step 2: Decide the significance level

α = 0.05

Step 3: Calculation of p-value

```python
from scipy.stats import spearmanr

# Dropping rows with NaN values for correlation analysis
df_cleaned_amount = df.dropna(subset=['Nominal GDP', 'default loan amount'])
df_cleaned_count = df.dropna(subset=['Nominal GDP', 'default loan count'])

# Calculating Spearmanr correlation coefficient and p-value for default loan amount
correlation_amount, p_value_amount = spearmanr(df_cleaned_amount['Nominal GDP'], df_cleaned_amount['default loan amount'])

# Calculating Spearmanr correlation coefficient and p-value for default loan count
correlation_count, p_value_count = spearmanr(df_cleaned_count['Nominal GDP'], df_cleaned_count['default loan count'])

# Printing the results
print("Correlation between  Nominal GDP and Default Loan Amount:", correlation_amount)
print("P-value for Default Loan Amount:", p_value_amount)

print("Correlation between  Nominal GDP and Default Loan Count:", correlation_count)
print("P-value for Default Loan Count:", p_value_count)
```

Results Interpretation:

- Correlation Coefficient: The correlation between Nominal GDP and both Default Loan Amount and Default Loan Count is 1.0. This indicates a perfect positive relationship, meaning that changes in Nominal GDP are perfectly aligned with changes in default loan amounts and counts.

-P-Value: Both p-values are 0.0, which indicates that the observed correlations are statistically significant.

Conclusion:

- The perfect correlation of 1.0 and p-values of 0.0 suggest a strong and statistically significant positive relationship between Nominal GDP and default loan amounts and counts. This means that as Nominal GDP increases, default loans also increase, and this relationship is highly significant.
